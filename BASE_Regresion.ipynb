**analsisi de los datos comportamiento con modificaciones**:

se enfoca en la limpieza e imputación de datos faltantes en un DataFrame llamado Data2f, particularmente en las columnas relacionadas con la producción de azúcar como "Ton.Azucar", "TonUltCorte", "Rdto", "TAH" y "%Sac.Caña". Inicialmente, los valores cero en "Ton.Azucar" se reemplazan por valores faltantes. Luego, se imputan los valores nulos en "Ton.Azucar" y "%Sac.Caña" utilizando la media de los valores dentro de grupos definidos por las columnas "Area Neta" y "Rdto" respectivamente, seguido de un llenado final con la media global para asegurar la ausencia de valores faltantes. Adicionalmente, se calculan o mantienen los valores de "TonUltCorte", "Ton.Azucar", "Rdto" y "TAH" mediante el uso de la función np.where, aplicando fórmulas específicas cuando los valores originales son nulos o cero, y conservando los valores existentes en caso contrario.

para esto ultimo se implementa las formulas
Las fórmulas para calcular las toneladas de azúcar extraída y otros parámetros en el cultivo de caña de azúcar provienen de estudios agronómicos, manuales de producción y literatura técnica en la industria azucarera. Aquí te doy las referencias clave de cada una:

**1. Fórmula para calcular las Toneladas de Azúcar Extraída** : Esta fórmula se basa en el concepto de **rendimiento sacaroso**, que mide cuánta azúcar se extrae en función del contenido de sacarosa de la caña.

Ton.Azúcar = Ton.Caña*( %Sac.Caña/100)

**Fuente:**  
- CENICAÑA (Centro de Investigación de la Caña de Azúcar de Colombia). Manuales de Producción de Caña y Procesamiento Industrial.  
- FAO (Organización de las Naciones Unidas para la Agricultura y la Alimentación). Informes sobre eficiencia en la producción de caña de azúcar.  
- CINCAE (Centro de Investigación de la Caña de Azúcar en Ecuador).  

**2. Fórmula del Rendimiento de Azúcar en Caña (Rdto o TAHM - Toneladas de Azúcar por Hectárea)** :Esta fórmula expresa el rendimiento en términos de porcentaje y es clave en estudios de eficiencia agrícola e industrial.

Rdto =( Ton.Azúcar/Ton.Caña)* 100

**Fuente:**  
- ICIDCA (Instituto Cubano de Investigaciones de los Derivados de la Caña de Azúcar). Estudios sobre la eficiencia del rendimiento agrícola y fabril.  
- Universidad de São Paulo, Brasil. Investigaciones sobre productividad y rendimiento sacaroso en variedades de caña de azúcar.  


3. Fórmula del TCH (Toneladas de Caña por Hectárea)** :Esta fórmula mide la productividad de la caña en términos de toneladas producidas por hectárea, un indicador esencial en la planificación agrícola.

TCH = Ton.Caña/Área Neta

**Fuente:**  
- CIRAD (Centro de Cooperación Internacional en Investigación Agronómica para el Desarrollo, Francia).  
- Manual de Producción de Caña de Azúcar de la **Sociedad Internacional de Técnicos de la Caña de Azúcar (ISSCT)**.  


```python
Data2f = data2.copy()
```

```python
print("Tamaño original (antes de eliminar nulos):", Data2f.shape)
missing_percentage = Data2f.isnull().mean()
columns_to_keep = missing_percentage[missing_percentage < 0.10].index
Data2f = Data2f[columns_to_keep]
print("Tamaño filtrado (después de eliminar columnas con muchos nulos):", Data2f.shape)
Data2f["Año_Corte"] = Data2f["F.Ult.Corte"].dt.year
Data2f["Mes_Corte"] = Data2f["F.Ult.Corte"].dt.month
Data2f["Año_Siembra"] = Data2f["F.Siembra"].dt.year
Data2f["Mes_Siembra"] = Data2f["F.Siembra"].dt.month
columnas_a_eliminar = ["Período", "Hacienda", "Cod.Estado", "Cod. T.Cultivo", "F.Siembra", "F.Ult.Corte"]
Data2f = Data2f.drop(columns=columnas_a_eliminar)
print("Tamaño filtrado (después de eliminar columnas especificadas):", Data2f.shape)
```

```python
for col in Data2f.select_dtypes(include=np.number):
    Data2f[col] = Data2f[col].round(3)
```

```python
Data2f['Rdto'].describe()
```

```python
percentile_50 = Data2f['Rdto'].quantile(0.5)
Data2f['Rdto'] = np.where(Data2f['Rdto'] == 0.0, percentile_50, Data2f['Rdto'])
percentile2_50 = Data2f['%Sac.Caña'].quantile(0.5)
Data2f['%Sac.Caña'] = np.where(Data2f['%Sac.Caña'] == 0.0, percentile2_50, Data2f['%Sac.Caña'])
percentile3_50 = Data2f['TAHM'].quantile(0.5)
Data2f['TAHM'] = np.where((Data2f['TAHM'] == 0.0) | (Data2f['TAHM'].isnull()), percentile3_50, Data2f['TAHM'])
Data2f['Ton.Azucar'] = np.where((Data2f['Ton.Azucar'].isnull()) | (Data2f['Ton.Azucar'] == 0), (Data2f['Rdto'] * Data2f['TonUltCorte']) / 100,
    Data2f['Ton.Azucar'])
Data2f['TAH'] = np.where((Data2f['TAH'].isnull()) | (Data2f['TAH'] == 0), (Data2f['Rdto'] * Data2f['TCH']) / 100,
    Data2f['TAH'])
Data2f['%Sac.Caña'] = np.where((Data2f['%Sac.Caña'].isnull()) | (Data2f['%Sac.Caña'] == 0), (Data2f['Ton.Azucar'] / Data2f['TonUltCorte']) * 100,
    Data2f['%Sac.Caña'])
Data2f = Data2f.fillna(0)
```

```python
# prompt: imputar  Ton.Azucar en base a los cuartiles 25, 50,75  de TCH, incluido los ceros de Ton.Azucar igualemte usar el valor del cualrtil correpodiente y en base a esos cuartiles asiganar los varoles usarlos colo limktes

# Calculate quartiles for TCH
quartiles_tch = Data2f['TCH'].quantile([0.25, 0.5, 0.75])

# Impute 'Ton.Azucar' based on TCH quartiles
for i in range(len(Data2f)):
    if Data2f['Ton.Azucar'][i] == 0:
        if Data2f['TCH'][i] <= quartiles_tch[0.25]:
            Data2f['Ton.Azucar'][i] = quartiles_tch[0.25]
        elif Data2f['TCH'][i] <= quartiles_tch[0.5]:
            Data2f['Ton.Azucar'][i] = quartiles_tch[0.5]
        elif Data2f['TCH'][i] <= quartiles_tch[0.75]:
            Data2f['Ton.Azucar'][i] = quartiles_tch[0.75]
        else:
            Data2f['Ton.Azucar'][i] = quartiles_tch[0.75] # Assign the value of the 75th percentile for values above the 75th percentile
```
